# Products By Year Launched

```python
df['Year Launched'].unique()
```

```python
df['Year Launched'].unique().min()
```

```python
df['Year Launched'].unique().max()
```

```python
product_by_year = df.groupby(['Year Launched'])['Product Name'].count().reset_index()
product_by_year
```

```python
plt.subplots(figsize=(12, 6))
plt.plot(product_by_year['Year Launched'],product_by_year['Product Name'],marker='o')
plt.xlabel('Year Launched')
plt.ylabel('Number of Products Launched')
plt.title('Number of Products Launched by Year')
plt.show()
```

# Products by Their Status

```python
df['Status'].unique()
```

```python
product_by_status = df.groupby(['Status'])['Product Name'].count().reset_index()
product_by_status
```

```python
fig, ax = plt.subplots()
ax.pie(product_by_status['Product Name'], labels=product_by_status['Status'],autopct='%1.1f%%', shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9}, startangle=90)
plt.title('Percentage of Products by Status')
plt.show()
```

# Products by Region Served

```python
df
```

```python
product_by_region = df.groupby(['Region Served'])['Product Name'].count().reset_index()
product_by_region
```

```python
plt.subplots(figsize=(12, 6))
plt.bar(product_by_region['Region Served'],product_by_region['Product Name'])
plt.title('Number of Products by Region Served')
plt.xlabel('Region Served')
plt.ylabel('Number of Products')
plt.xticks(rotation=90)
plt.show()
```

```python
product_by_region['Percentage of Total Products'] = (product_by_region['Product Name']/product_by_region['Product Name'].sum())*100
product_by_region
```

# Product Analysis Based On Their Status (Active/Discontinued)

```python
active_products = df[df['Status']=='Active']
active_products
```

```python
active_products_year = active_products.groupby(['Year Launched'])['Product Name'].count().reset_index()
active_products_year
```

```python
discontinued_products = df[df['Status']=='Discontinued']
discontinued_products
```

```python
Discontinued_products_year = discontinued_products.groupby(['Year Launched'])['Product Name'].count().reset_index()
Discontinued_products_year
```

```python
plt.subplots(figsize=(12, 6))
plt.plot(Discontinued_products_year['Year Launched'],Discontinued_products_year['Product Name'],marker='o')
plt.xlabel('Year Launched')
plt.ylabel('Number of Products Discontinued')
plt.title('Number of Discontinued Products by Year')
plt.show()
```

```python
plt.subplots(figsize=(12, 6))
plt.plot(active_products_year['Year Launched'],active_products_year['Product Name'],marker='o')
plt.xlabel('Year Launched')
plt.ylabel('Number of Active Products')
plt.title('Number of Active Products by Year')
plt.show()
```
